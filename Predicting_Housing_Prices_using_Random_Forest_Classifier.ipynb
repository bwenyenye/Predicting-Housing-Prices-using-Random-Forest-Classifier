**Loading Zillow House Price Data**

```python
zillow_files = ["/content/Metro_median_sale_price_uc_sfrcondo_week.csv", "/content/Metro_zhvi_uc_sfrcondo_tier_0.33_0.67_month.csv"]

dfs = [pd.read_csv(f) for f in zillow_files]
```

Each row contains data about a region. The first row contains the median sales price for the whole country. We will focus on that here.

Sales price dataset comes out weekly, while our house value data only comes out monthly.We need to combine them.The loop below will cut out the date column and remain with the month and year columns giving us a common column to join the two datasets on.

```python
for df in dfs :
  df.index = pd.to_datetime(df.index)
  df["month"] = df.index.to_period("M") 
```

Both datasets now have a month column. We use the month column to combine these dataframes.

```python
price_data = dfs[0].merge(dfs[1], on ="month")
```

Making sure that the index of this dataframe is the same as the index of our first dataframe.

```python
price_data.index = dfs[0].index
```

```python
price_data
```

Changing our column names and deleting the month column since we dont need it anymore

```python
del price_data["month"]
price_data.columns = ["price", "value"]
```

**Preparing Data for Machine Learning**

We need to merge this Zillow price data with our Federal Reserve economic data.
To do so, we will start by dropping the missing values in fed data

```python
fed_data = fed_data.dropna()
```

```python
fed_data
```

We need to align our fed data with our zillow data in terms of dates.They are about 2 dates off each other.To do so, we are going to add a couple of dates to our federal reserve data.

```python
from datetime import timedelta
fed_data.index = fed_data.index + timedelta(days=2)
```

```python
fed_data
```

You can see everything has been shifted foward by 2 days. We are now able to merge it with the zillow data because the dates are consistent.

```python
price_data = fed_data.merge(price_data, left_index=True, right_index=True)
```

Renaming columns


```python
price_data.columns =["Interest", "vacancy", "cpi", "price", "value"]
```

**Data Exploration**
**Setting up our machine learning target**

```python
price_data.plot(y="price", use_index = True)
```

```python
price_data["adj_price"] = price_data["price"] / price_data["cpi"] * 100
```

```python
price_data.plot.line(y="adj_price", use_index=True)
```

```python
price_data["adj_value"] = price_data["value"] / price_data["cpi"] * 100
```

The shift methods grabs the adjusted price from 3 months into the future and pulls it back to the current row.

```python
price_data["next_quarter"] = price_data["adj_price"].shift(-13)
```

```python
price_data
```

```python
price_data.dropna(inplace=True)
```

```python
price_data
```

```python
price_data["change"] = (price_data["next_quarter"] > price_data["adj_price"]).astype(int)
```

```python
price_data
```

```python
price_data["change"].value_counts()
```

```python
predictors = ["interest", "vacancy", "adj_price", "adj_value"]
target = "change"
```

**Creating a machine learning model**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import numpy as np
```

**Creating a Backtessing Engine**

```python
START = 260
STEP = 52

def predict(train, test, predictors, target):
    rf = RandomForestClassifier(min_samples_split=10, random_state=1)
    rf.fit(train[predictors], train[target])
    preds = rf.predict(test[predictors])
    return preds

def backtest(data, predictors, target):
    all_preds = []
    for i in range(START, data.shape[0], STEP):
        train = price_data.iloc[:i]
        test = price_data.iloc[i:(i+STEP)]
        all_preds.append(predict(train, test, predictors, target))
    
    preds = np.concatenate(all_preds)
    return preds, accuracy_score(data.iloc[START:][target], preds)
```

```python
preds, accuracy = backtest(price_data, predictors, target)
```
